```python
# Assuming you have already replaced "Sim" and "Não" with 1 and 0
results[["Gostaste da imagem #1 recomendada?", "Gostaste da imagem #2 recomendada?", "Gostaste da imagem #3 recomendada?", "Gostaste da imagem #4 recomendada?", "Gostaste da imagem #5 recomendada?"]] = results[["Gostaste da imagem #1 recomendada?", "Gostaste da imagem #2 recomendada?", "Gostaste da imagem #3 recomendada?", "Gostaste da imagem #4 recomendada?", "Gostaste da imagem #5 recomendada?"]].replace({'Sim': 1, 'Não': 0})

# Function to print Precision in percentage
def print_precision(df, system_type, system_name):
    sys = df[df["Tipo do Sistema"] == system_type]
    
    print(f"--- {system_name} ---")
    
    # Precision@1
    precision_1 = sys["Gostaste da imagem #1 recomendada?"].sum() / len(sys)
    print(f"Precision@1 = {precision_1:.3f}")
    5
    # Precision@2
    precision_2 = sys[["Gostaste da imagem #1 recomendada?", "Gostaste da imagem #2 recomendada?"]].sum().sum() / (2 * len(sys))
    print(f"Precision@2 = {precision_2:.3f}")
    
    # Precision@3
    precision_3 = sys[["Gostaste da imagem #1 recomendada?", "Gostaste da imagem #2 recomendada?", "Gostaste da imagem #3 recomendada?"]].sum().sum() / (3 * len(sys))
    print(f"Precision@3 = {precision_3:.3f}")
    
    # Precision@4
    precision_4 = sys[["Gostaste da imagem #1 recomendada?", "Gostaste da imagem #2 recomendada?", "Gostaste da imagem #3 recomendada?", "Gostaste da imagem #4 recomendada?"]].sum().sum() / (4 * len(sys))
    print(f"Precision@4 = {precision_4:.3f}")
    
    # Precision@5
    precision_5 = sys[["Gostaste da imagem #1 recomendada?", "Gostaste da imagem #2 recomendada?", "Gostaste da imagem #3 recomendada?", "Gostaste da imagem #4 recomendada?", "Gostaste da imagem #5 recomendada?"]].sum().sum() / (5 * len(sys))
    print(f"Precision@5 = {precision_5:.3f}")

# Print for System 1
print_precision(results, 1, "CBF")

# Print for System 2
print_precision(results, 2, "CF")

# Print for System 3
print_precision(results, 3, "RANDOM")
```

```python
results[["Apresenta o teu ranking pessoal das imagens recomendadas [Imagem #1]", "Apresenta o teu ranking pessoal das imagens recomendadas [Imagem #2]", "Apresenta o teu ranking pessoal das imagens recomendadas [Imagem #3]", 
         "Apresenta o teu ranking pessoal das imagens recomendadas [Imagem #4]", "Apresenta o teu ranking pessoal das imagens recomendadas [Imagem #5]"]] = results[["Apresenta o teu ranking pessoal das imagens recomendadas [Imagem #1]", 
        "Apresenta o teu ranking pessoal das imagens recomendadas [Imagem #2]", "Apresenta o teu ranking pessoal das imagens recomendadas [Imagem #3]", "Apresenta o teu ranking pessoal das imagens recomendadas [Imagem #4]", 
        "Apresenta o teu ranking pessoal das imagens recomendadas [Imagem #5]"]].replace({'1º lugar': 1, '2º lugar': 2, '3º lugar': 3, '4º lugar': 4, '5º lugar': 5})

def print_rank(df, system_type, system_name):
    sys = df[df["Tipo do Sistema"] == system_type]
    
    print(f"--- {system_name} ---")
    # Média dos rankings por posição de recomendação
    for i in range(1, 6):
        media_rank = sys[f"Apresenta o teu ranking pessoal das imagens recomendadas [Imagem #{i}]"].mean()
        print(f"Média do Ranking da Imagem #{i}: {media_rank:.2f}")

# Print for System 1
print_rank(results, 1, "CBF")

# Print for System 2
print_rank(results, 2, "CF")

# Print for System 3
print_rank(results, 3, "RANDOM")
```
