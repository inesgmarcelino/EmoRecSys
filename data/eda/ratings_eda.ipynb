```python
# ---- IMPORTS ---- #
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

dataset_cf = pd.read_csv("../csvs/ratings.csv")
```

```python
dataset_cf.describe(include='all') # -- este!
```

```python
 # -- este!
for col in dataset_cf.columns[3:]: # categorical variables 
    plt.figure(figsize=(10,6))
    sns.countplot(data=dataset_cf, x=col, order=dataset_cf[col].value_counts().index)
    # plt.title(f'Distribution of {col} variable')
    # plt.xticks(rotation=90)
    # plt.tight_layout()
    plt.show()
```

```python
user_interactions = dataset_cf['id_survey'].value_counts()
item_interactions = dataset_cf['id_photo'].value_counts()

print("Nmero mdio de interaes por usurio:", user_interactions.mean())
print("Nmero mdio de interaes por item:", item_interactions.mean())
print("Nmero mnimo de interaes por usurio:", user_interactions.min())
print("Nmero mnimo de interaes por item:", item_interactions.min())
```

```python
plt.figure(figsize=(12, 6))
plt.hist(user_interactions, bins=50, alpha=0.5, label='Interaes por Usurio')
plt.hist(item_interactions, bins=50, alpha=0.5, label='Interaes por Item')
plt.legend(loc='upper right')
plt.xlabel('Nmero de Interaes')
plt.ylabel('Frequncia')
plt.title('Distribuio de Interaes')
plt.show()
```
