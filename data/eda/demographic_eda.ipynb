### Univariate Analysis

#### Distribution of Categorical Variables

```python
max_y = len(dataset_demo) 
for col in dataset_demo.columns[4:7]: # categorical variables 
    plt.figure(figsize=(10,6))
    sns.countplot(data=dataset_demo, x=col, order=dataset_demo[col].value_counts().index)
    plt.ylim(0, max_y)  # Definindo o mesmo limite para o eixo Y
    plt.yticks(fontsize=13)
    plt.xticks(rotation=90, fontsize=13)  # Ajustando a rotação e o tamanho das labels do eixo X
    plt.xlabel(col, fontsize=16)
    plt.ylabel('Number of Participants', fontsize=16)
    plt.tight_layout()
    plt.show()
```

#### Distribution of Binary Categorical Variables

```python
for col in dataset_demo.columns[10:]:
    plt.figure(figsize=(10,6))
    sns.countplot(x=col, data=dataset_demo)
    plt.title(f'Distribution of {col} variable')
    plt.xticks([0, 1], ['Absent', 'Present'])
    plt.tight_layout()
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=dataset_demo, x='age', order=dataset_demo['age'].value_counts().index)
plt.title('Distribution of Age')
plt.savefig('age_distribution.png', dpi=300)
```
