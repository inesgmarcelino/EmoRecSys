### Univariate Analysis

#### Distribution of Categorical Variables

```python
dataset_demo['gender'] = dataset_demo['gender'].replace({'F': 'Female', 'M': 'Male', 'O': 'Other'})
dataset_demo['gender']
```

```python
max_y = len(dataset_demo)

for col in dataset_demo.columns[1:4]:  # categorical variables
    plt.figure(figsize=(10, 6))
    
    # Criando o gráfico de barras
    ax = sns.countplot(
        data=dataset_demo,
        x=col,
        order=dataset_demo[col].value_counts().index,
        color="#92befc"
    )

    # Calculando os valores absolutos e percentuais
    total = len(dataset_demo)
    for p in ax.patches:
        count = p.get_height()  # Altura da barra
        percentage = 100 * count / total  # Calculando a percentagem
        ax.annotate(
            f'{percentage:.1f}%',  # Formatando a percentagem
            (p.get_x() + p.get_width() / 2, count),  # Posição do texto
            ha='center', va='bottom', fontsize=12, color='black'  # Estilo do texto
        )

    # Configurando o eixo Y e outros elementos
    plt.ylim(0, max_y)
    plt.yticks(fontsize=13)
    plt.xticks(rotation=0, fontsize=13)
    plt.xlabel(col, fontsize=16)
    plt.ylabel('Number of Participants', fontsize=16)
    plt.tight_layout()

    # Salvando o gráfico
    plt.savefig(f"{col}_distribution.svg", format="svg")
    plt.show()
```

```python
max_y = len(dataset_demo)

for col in dataset_demo.columns[6:7]:  # categorical variables
    plt.figure(figsize=(10, 6))

    value_counts = dataset_demo[col].value_counts()
    top_11 = value_counts.nlargest(5)
    top_11_names = top_11.index
    dataset_demo[col] = dataset_demo[col].apply(lambda x: x if x in top_11_names else "Other")
    
    # Criando o gráfico de barras
    ax = sns.countplot(
        data=dataset_demo,
        x=col,
        order=dataset_demo[col].value_counts().index,
        color="#92befc"
    )

    # Calculando os valores absolutos e percentuais
    total = len(dataset_demo)
    for p in ax.patches:
        count = p.get_height()  # Altura da barra
        percentage = 100 * count / total  # Calculando a percentagem
        ax.annotate(
            f'{percentage:.1f}%',  # Formatando a percentagem
            (p.get_x() + p.get_width() / 2, count),  # Posição do texto
            ha='center', va='bottom', fontsize=12, color='black'  # Estilo do texto
        )

    # Configurando o eixo Y e outros elementos
    plt.ylim(0, max_y)
    plt.yticks(fontsize=13)
    plt.xticks(rotation=0, fontsize=13)
    plt.xlabel(col, fontsize=16)
    plt.ylabel('Number of Participants', fontsize=16)
    plt.tight_layout()

    # Salvando o gráfico
    plt.savefig(f"{col}_distribution.svg", format="svg")
    plt.show()
```

```python
hobbies = {"Crafts": 17,
           "Hiking": 34,
           "Cycling": 13,
           "Cinema": 31,
           "Running": 11,
           "Cooking": 22,
           "Dancing": 7,
           "Drawing": 16,
           "Sculpture": 1,
           "Photography": 13,
           "Gardening": 13,
           "Board Games": 38,
           "Video Games": 47, 
           "Reading": 49,
           "Music": 46,
           "Swimming": 4,
           "Painting": 9,
           "Theater": 5}

sorted_hobbies = sorted(hobbies.items(), key=lambda x: x[1], reverse=True)
top_7 = dict(sorted_hobbies[:10])
others = sum(value for _, value in sorted_hobbies[10:])

# Adicionando "Others" aos dados
top_7["Others"] = others

# Dados para o gráfico
labels = list(top_7.keys())
sizes = list(top_7.values())

colors = ["#92befc", "#261CE6", "#6CA6FE", "#3F89FD", "#85E3FF", 
          "#92E6E6", "#C7CEFF", "#99A6FF", "#5874E6", "#4056E6", "#AEC6FF"]

# Criando o gráfico de pizza
plt.figure(figsize=(10, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, textprops={'fontsize': 16}, colors=colors)
plt.tight_layout()
plt.savefig(f"hobbies_distribution.svg", format="svg")
plt.show()
```
